```python
# Convertendo datas

# Transformando em um tipo de dado datetime
df["created_at"] = pd.to_datetime(df["created_at"])

```

```python
# Realizando filtros: 

df_filtrado = df[(df["created_at"].dt.year >= 2018) & (df["created_at"].dt.year <= 2018)]
df_filtrado.head()
```

```python
df_filtrado2 = df[(df["created_at"].dt.year == 2018) & (df["created_at"].dt.month == 1)]
df_filtrado2
```

```python
# Transformando a data no formato brasileiro

df_filtrado2["created_at"] = df_filtrado2["created_at"].dt.strftime("%d/%m/%y")
```

```python
# Adicionando colunas de ano, mês e dia.
df["created_at"] = pd.to_datetime(df["created_at"]) #Convertendo para hora o tipo de dado.
df["ano"] = df["created_at"].dt.year #Adiciona a coluna com o ano
df["mês"] = df["created_at"].dt.month #Adiciona a coluna com o mês
df["dia"] = df["created_at"].dt.day #Adiciona a coluna com o dia
meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]
df["NomeMês"] = df["created_at"].dt.month.map(lambda x: meses[x-1])
df
```
